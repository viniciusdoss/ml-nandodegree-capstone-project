#### Separating Training and Test data:

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# Get features and labels
X = df.drop(['Class'], axis=1).values
y = df['Class'].values
```

* Here we apply a cross-validation hold-out using 20% of the data for testing the model and 80% for effective training:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
print(X_train.shape, '-->', y_train.shape)
print(X_test.shape, '-->', y_test.shape)
```

* Now we scale features:

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

* Now let's create a function with some Machine Learning models that we want to test:

```python
from sklearn.metrics import recall_score, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

def test_models(X_train, y_train):
    
    # Logistic Regression
    reglog = LogisticRegression(random_state = 7)
    reglog.fit(X_train, y_train)
    
    # Decision Tree
    tree = DecisionTreeClassifier(criterion = 'entropy', random_state = 7)
    tree.fit(X_train, y_train)

    # Random Forest
    forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 7)
    forest.fit(X_train, y_train)
    
    # KNeighborsClassifier (KNN)
    knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
    knn.fit(X_train, y_train)

    # SVM Support Vector Machine with linear kernel
    linear_svm = SVC(kernel = 'linear', random_state = 7, probability=True)
    linear_svm.fit(X_train, y_train)
 
    # SVM Support Vector Machine with radial basis kernel 
    rbf_svm = SVC(kernel = 'rbf', random_state = 7, probability=True)
    rbf_svm.fit(X_train, y_train)


    # Show accuracy and recall of all models
    print('0 - Logistic Regression:    {} of accuracy and {} of recall'.format(reglog.score(X_train, y_train), recall_score(y_train, reglog.predict(X_train))))
    print('1 - Decision Tree:          {} of accuracy and {} of recall'.format(tree.score(X_train, y_train), recall_score(y_train, tree.predict(X_train))))
    print('2 - Random Forest:          {} of accuracy and {} of recall'.format(forest.score(X_train, y_train), recall_score(y_train, forest.predict(X_train))))
    print('3 - KNN:                    {} of accuracy and {} of recall'.format(knn.score(X_train, y_train), recall_score(y_train, knn.predict(X_train))))
    print('4 - SVM with linear kernel: {} of accuracy and {} of recall'.format(linear_svm.score(X_train, y_train), recall_score(y_train, linear_svm.predict(X_train))))
    print('5 - SVM with rbf kernel:    {} of accuracy and {} of recall'.format(rbf_svm.score(X_train, y_train), recall_score(y_train, rbf_svm.predict(X_train))))

    return reglog, tree, forest, knn, linear_svm, rbf_svm
```

* Training models with metrics in the training data:

```python
model = test_models(X_train, y_train)
```

```python
model
```

* Now we are going to generate our Precision-Recall Curve for each model:

```python
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score, auc
```

```python
model_names_list = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'KNN', 'SVM with linear kernel', 'SVM with rbf kernel']
```

```python
y_test
```

```python
for m, m_name in zip(model, model_names_list):
    # precision-recall curve and f1

    # predict probabilities
    lr_probs = m.predict_proba(X_test)
    
    # keep probabilities for the positive outcome only
    lr_probs = lr_probs[:, 1]
    
    # predict class values
    yhat = m.predict(X_test)
    
    
    lr_precision, lr_recall, _ = precision_recall_curve(y_test, lr_probs)
    
    lr_f1, lr_auc = f1_score(y_test, yhat), auc(lr_recall, lr_precision)
    
    
    # summarize scores
    print(m_name, ': f1=%.3f auc=%.3f' % (lr_f1, lr_auc))
    
    
    # plot the precision-recall curves
    no_skill = len(y_test[y_test==1]) / len(y_test)
    plt.plot([0, 1], [no_skill, no_skill], linestyle='--', label='No Skill')
    plt.plot(lr_recall, lr_precision, marker='.', label=m_name)
    # axis labels
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    # show the legend
    plt.legend()
    # show the plot
    plt.show()
```

* Note that in the training data the best accuracy accompanied by the best recall was the decision tree model and in the test datan the random forest model has the best Precision-Recall Curve área (auc = 0.866). That indicates that Random Forest in default configuration was the best model.
